```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import mglearn

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Gerando dados
X, y = make_blobs(random_state=1)
```

```python
# Visualizando os dados
plt.scatter(X[:, 0], X[:, 1], s=150, linewidths=1.5, edgecolors='black')
plt.xlabel('NÂº Compras')
plt.ylabel('Qtd de Produtos')
plt.show()
```

```python
# Criando o modelo
model_kmeans = KMeans(n_clusters=3, random_state=0)
model_kmeans.fit(X)

labels = model_kmeans.predict(X)
centroids = model_kmeans.cluster_centers_
```

```python
# Visualizando os clusters formados
mglearn.plots.plot_kmeans_algorithm()

compras = np.random.uniform(-12, 0, 5)
produtos = np.random.uniform(-10, 5, 5)
X_test = np.column_stack((compras, produtos))
```

```python
# Visualizando o algoritmo com suas fronteiras delimitadas
mglearn.plots.plot_kmeans_boundaries()
```
