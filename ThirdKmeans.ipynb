```python
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
# Carregando a base de dados iris
iris = datasets.load_iris()
iris
```

```python
def plot_clusters(data, labels, title):
    colors = ['red', 'green', 'purple', 'black']
    plt.figure(figsize=(8, 4))
    for i,c,l in zip(range(-1, 3), colors, ['Noise', 'Setosa', 'Versicolor', 'Virginica']):
        if i == -1:
            plt.scatter(data[labels == i, 0], data[labels == i, 3], c=colors[i], label = l, alpha = 0.5, s = 50, marker = 'x')
        else:
            plt.scatter(data[labels == i, 0], data[labels == i, 3], c=colors[i], label = l, alpha = 0.5, s = 50)
    plt.legend()
    plt.title(title)
    plt.xlabel('Comprimento da Sépala')
    plt.ylabel('Largura da Pétala')
    plt.show()
```

```python
# Verificando os dados
iris.target
```

```python
# Utilizando o KMeans
kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(iris.data) #treinando o modelo
print(kmeans.labels_)
resultados = confusion_matrix(iris.target, kmeans.labels_) #comparando os resultados com o método confusion_matrix
print(resultados)

plot_clusters(iris.data, kmeans.labels_, 'Cluster KMeans')
```

```python
#dps = circunferencia vai buscar os pontos
#min_samples = número mínimo de pontos para formar um cluster
#-1 seria ruído
dbscan = DBSCAN(eps=0.5, min_samples=3)
dbscan_labels = dbscan.fit_predict(iris.data)
print(dbscan_labels)

plot_clusters(iris.data, dbscan_labels, 'Cluster DBSCAN')
```

```python
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(iris.data)
print(agglo_labels)
plot_clusters(iris.data, agglo_labels, 'Cluster Agglomerative')
```

```python
# Ward é uma função que calcula a distância entre os clusters
plt.figure(figsize=(12, 6))
plt.title('Cluster Hierárquico: Dendrograma')
plt.xlabel('Índice')
plt.ylabel('Distância')
linkage_matrix = linkage(iris.data, method='ward')
dendrogram(linkage_matrix, truncate_mode='lastp', p=15)
plt.axhline(y=7, c='gray', linestyle = 'dashed')
plt.show()
```
